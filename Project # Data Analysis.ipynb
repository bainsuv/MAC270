```python
Mac270=pd.read_table('chipotle.tsv')
Mac270.head(5)
#Import the dataset from file on blackboard
#First you need to download the tsv file and save it in your computer. Second, you
#need to upload the file in jupyter notebook as mentioned before as a variable
#named MAC270.
```

```python
itemByQuantity = Mac270.groupby('item_name')['quantity'].sum()
items_table = pd.DataFrame.from_dict(itemByQuantity)
items_table_sorted = items_table.sort_values(by=['quantity'], ascending=False)
items_table_sorted.head()
ax = items_table_sorted.iloc[:5].plot.bar(y='quantity', legend=False)
ax.set_xlabel('item_name')
ax.set_ylabel('quantity')
ax.set_title('Top 5 items bought')
#Assign your data to a variable called MAC270.
#Show the first 5 rows
#.Create a histogram of the top 5 items bought
```

```python
itemPerPrice = Mac270.groupby('item_price')['quantity'].sum()
price_table = pd.DataFrame.from_dict(itemPerPrice)
price_table['index'] = price_table.index
price_table['index'] = price_table['index'].map(lambda x: x.lstrip('$').rstrip
('aAbBcC'))
price_table['index'] = price_table['index'].astype(float)
price_table = price_table.sort_values('index', ascending=False)
price_table['index'].head()
plt.figure(figsize=(5,25))
plt.scatter(price_table.quantity, price_table.index)
plt.ylabel('Price', fontsize=25)
plt.xlabel('Quantity', fontsize=25)
plt.title('Quantity per price', fontsize=25)
#Create a scatterplot with the number of items ordered per order price
```
